```python
import torch
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/Churn_Modelling.csv")
df
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

df['val1']=df['Geography'].apply(lambda x: 1 if x=='France' else 2 if x=='Germany' else 3  )
df['val2']=df['Gender'].apply(lambda x: 1 if x=='Male' else 2  )
df


```

```python
a= df['Exited'].values
b=df.drop(['RowNumber','Surname','Geography','Gender','Exited'],axis=1).values

X_train,X_test,y_train,y_test=train_test_split(b,a, test_size=0.2)
```

```python
X_train
```

```python
x= StandardScaler()
X_train =x.fit_transform(X_train)
X_test =x.fit_transform(X_test)
X_train

```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
class exited(nn.Module):

  def __init__(self):
      super(exited, self).__init__()
      self.l1 = nn.Linear(11,100)
      self.l2 = nn.Linear(100,100)
      self.output =nn.Linear(100,1)

  def forward(self,x):
      x=F.relu(self.l1(x))
      x=F.sigmoid(self.l2(x))
      x=torch.sigmoid(self.output(x))
      return x
```

```python
model = exited()
criterion = nn.BCELoss() # Binary Cross Entropy Loss
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

epochs = 200
for epoch in range(epochs):
    # Convert arrays to tensors
    inputs = torch.tensor(X_train, dtype=torch.float32)
    labels = torch.tensor(y_train, dtype=torch.float32)

    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, labels.unsqueeze(1))

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```

```python
with torch.no_grad():
    y_predicted = model(torch.tensor(X_test, dtype=torch.float32))
    y_predicted_cls = y_predicted.round()
    acc = y_predicted_cls.eq(torch.tensor(y_test).unsqueeze(1)).sum()/ float(y_test.shape[0])
    print(f'Accuracy: {acc:.4f}')

```
