```python
# Importing required libraries

import nltk  # natural language toolkit
import pandas as pd # a library that handles big dataframes
from nltk.corpus import stopwords # importing a set of stopwords to clean the data and remove unneccesary words for better processing
from textblob import Word #importing the word object for lemmetisation
from sklearn.preprocessing import LabelEncoder
from collections import Counter
import wordcloud
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("lakshmi25npathi/imdb-dataset-of-50k-movie-reviews")

print("Path to dataset files:", path)
```

```python
import os

# List the contents of the downloaded directory
files_in_path = os.listdir(path)
print(files_in_path)
```

```python
import os

df = pd.read_csv(os.path.join(path, 'IMDB Dataset.csv'))
df.head()
```

```python
data_v1 = df[['review','sentiment']] #checing for null values in the dataset
data_v1.isnull().sum()
```

```python
def cleaning(df, stop_words):
    # this basically splits the sentence into set of words and then makes it lowercase after that it outputs it as a string of lowercase words with single space
    df['review'] = df['review'].apply(lambda x:' '.join(x.lower() for x in x.split()))

    # Replacing the special characters with space
    df['review'] = df['review'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)

    # Replacing the digits/numbers left after cleaning the text
    df['review'] = df['review'].str.replace(r'\d','', regex=True)

    # Removing stop words
    df['review'] = df['review'].apply(lambda x:' '.join(x for x in x.split() if x not in stop_words))

    # Lemmatization
    df['review'] = df['review'].apply(lambda x:' '.join([Word(x).lemmatize() for x in x.split()]))#  here we first convert words in x.split to textblob word object then lemmetize it

    return df

import nltk
nltk.download('stopwords')
nltk.download('wordnet') # Download wordnet for lemmatization

stop_words = stopwords.words('english')

data_v1 = cleaning(data_v1, stop_words)
print(df['review'][0])
print(data_v1.review[0])
```

```python
#creating a vocabulary of all the words for analysing it later
common_words=''
for i in data_v1.review:
    i = str(i) # converting each review to string
    tokens = i.split() #outputs a array of all the words in a given review separated by spaces
    common_words += " ".join(tokens)+" "#here the vocabulary is created by adding each word to the common word string
wordcloud = wordcloud.WordCloud().generate(common_words)#Creates a WordCloud object with default settings and then parses the whole string and counts frequency of each word
plt.imshow(wordcloud, interpolation='bilinear')#bilinear smooths out pixels
plt.axis("off")#turning of axes for better visualisation
plt.show()
```

```python
# Encoded the target column
lb=LabelEncoder()
data_v1['sentiment'] = lb.fit_transform(data_v1['sentiment'])
```

```python
tokenizer = Tokenizer(num_words=500, split=' ')
tokenizer.fit_on_texts(data_v1['review'].values)
X = tokenizer.texts_to_sequences(data_v1['review'].values)
X = pad_sequences(X)
```

```python
from keras.layers import Input

model = Sequential()
model.add(Input(shape=(X.shape[1],))) # Dynamically set input shape to match padded sequence length
model.add(Embedding(500,120))
model.add(SpatialDropout1D(0.4))
model.add(LSTM(176,dropout=0.2,recurrent_dropout=0.2))
model.add(Dense(1,activation='sigmoid'))

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
#Splitting the data into training and testing
y_labels = data_v1['sentiment'] # sentiment is already 0/1 after LabelEncoder
X_train, X_test, y_train, y_test = train_test_split(X, y_labels, test_size = 0.3, random_state = 42)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Step 10: Training with early stopping
batch_size = 256

es = EarlyStopping(
    monitor='val_loss',           # Watch validation loss
    patience=10,                   # Stop after 5 epochs no improvement
    restore_best_weights=True,    # Return to best weights
    verbose=1                     # Print when stopping
)

model.fit(X_train, y_train,
          epochs=200,
          batch_size=batch_size,
          validation_split=0.2,     # Use 20% of training data as validation
          callbacks=[es],           # Apply early stopping
          verbose='auto')
```
