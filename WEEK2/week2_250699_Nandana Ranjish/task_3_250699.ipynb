```python
# loading a sentiment analyzer

senti=pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
result=senti("Mission Impossible series is a really dedicated work of Tom Cruise")
print(result)
```

```python
# loading text generator and prompting

text_gen=pipeline('text-generation', model="gpt2")
prompt="Explain what is data preprocessing."
result=text_gen(prompt, max_length=300, num_return_sequences=1)
print(result[0]["generated_text"])
```

```python
qa=pipeline('question-answering', model='distilbert-base-cased-distilled-squad')
context="The transformer is a neural network with a specific structure that includes a mechanism called self-attention or multi-head attention."
ques="What mechanism does a transformer include?"
result=qa(question=ques, context=context)
print(result)
```
