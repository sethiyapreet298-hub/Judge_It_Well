## Check Matrix Multiplication Validity

### Subtask:
Define two matrices and write code to determine if matrix multiplication is valid by comparing their dimensions (number of columns in the first matrix must equal the number of rows in the second matrix).


**Reasoning**:
The subtask requires defining two matrices and checking if their dimensions allow for matrix multiplication. This code block will define the matrices, extract their dimensions, and perform the validity check as instructed.



```python
matrix1 = [[1, 2, 3], [4, 5, 6]] # 2x3 matrix
matrix2 = [[7, 8], [9, 10], [11, 12]] # 3x2 matrix

rows1 = len(matrix1)
cols1 = len(matrix1[0])

rows2 = len(matrix2)
cols2 = len(matrix2[0])

print(f"Matrix 1 dimensions: {rows1}x{cols1}")
print(f"Matrix 2 dimensions: {rows2}x{cols2}")

if cols1 == rows2:
    print("Matrix multiplication is valid.")
else:
    print("Matrix multiplication is not valid.")

# Example of incompatible matrices
print("\n--- Testing incompatible matrices ---")
matrix3 = [[1, 2], [3, 4]] # 2x2 matrix
matrix4 = [[5, 6, 7], [8, 9, 10]] # 2x3 matrix (compatible, just for example)
matrix5 = [[1,2,3],[4,5,6]] # 2x3 matrix
matrix6 = [[7,8],[9,10]] # 2x2 matrix

rows5 = len(matrix5)
cols5 = len(matrix5[0])

rows6 = len(matrix6)
cols6 = len(matrix6[0])

print(f"Matrix 5 dimensions: {rows5}x{cols5}")
print(f"Matrix 6 dimensions: {rows6}x{cols6}")

if cols5 == rows6:
    print("Matrix multiplication is valid.")
else:
    print("Matrix multiplication is not valid.")
```

## Matrix Multiplication using Loops

### Subtask:
Implement matrix multiplication for the defined matrices using nested loops, without relying on the NumPy library. This involves iterating through rows, columns, and the inner product dimension.


**Reasoning**:
The subtask requires implementing matrix multiplication using nested loops. This code block will initialize the result matrix and then perform the multiplication as specified in the instructions, without using NumPy.



```python
if cols1 == rows2:
    # Initialize result matrix with zeros
    result_matrix = [[0 for _ in range(cols2)] for _ in range(rows1)]

    # Perform matrix multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

    print("Result of matrix multiplication:")
    for row in result_matrix:
        print(row)
else:
    print("Matrix multiplication is not valid for the given matrices.")
```
