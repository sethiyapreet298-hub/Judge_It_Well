## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 100)
y = x**2

y_for_plot = 2 * x


fig = plt.figure()


ax = fig.add_axes([0, 0, 1, 1])


ax.plot(x, y_for_plot, color='blue')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')


plt.show()
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 100)

y = 2 * x


fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])


ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])


ax1.plot(x, y, color='blue')
ax1.set_xlabel('X-axis Large')
ax1.set_ylabel('Y-axis Large')
ax1.set_title('Main Plot')


ax2.plot(x, y, color='red')
ax2.set_xlabel('X-axis Small')
ax2.set_ylabel('Y-axis Small')
ax2.set_title('Inset Plot')

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 100)
z = x**2

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x, z, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_xlim([0, 100])
ax1.set_ylim([0, 10000])

ax2.plot(x, z, color='blue')
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python

fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, color='blue', linestyle='--')
axes[0].set_xlim([0, 100])
axes[0].set_ylim([0, 200])


axes[1].plot(x, z, color='red', linewidth=2)
axes[1].set_xlim([0, 100])
axes[1].set_ylim([0, 10000])

```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
axes[0].plot(x, y, color='blue', linewidth=3)
axes[1].plot(x, z, color='red', linestyle='--')
plt.show()
```

# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]

# 1. Histogram
plt.figure()
plt.hist(data, bins=30)
plt.title('Histogram Example')
plt.show()

# 2. Scatter Plot
plt.figure()
plt.scatter(data[:100], data[100:200])
plt.title('Scatter Plot Example')
plt.show()

# 3. Boxplot
plt.figure()
plt.boxplot([data[:250], data[250:500]])
plt.title('Boxplot Example')
plt.show()

# 4. Pie charts
plt.figure()
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```
