## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import numpy as np
x = np.arange(0,100)
y = x*2
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,z)
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x,y, color='blue', lw=3, ls='--')
axes[1].plot(x,z, color='red', lw=4, ls='-')
fig
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

axes[0].plot(x,y, color='blue', lw=3, ls='--')
axes[1].plot(x,z, color='red', lw=4, ls='-')
fig
```

# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","C","D","E","F"])
y = np.array([3,4,6,7,9,12])
z = np.array([6,9,2,15,6,7])

plt.bar(x,y)
plt.show()

plt.hist(y)
plt.show()

plt.pie(y)
plt.show()

plt.scatter(y,z)
plt.show()

plt.boxplot(y)
plt.show()
```
