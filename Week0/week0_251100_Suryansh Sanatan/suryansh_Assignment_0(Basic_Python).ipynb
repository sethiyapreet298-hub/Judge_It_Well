(i) Given:

```python
X = [1,2,3,4,5]
Y = [2,4,5,4,5]
sumX=0
sumY=0
for i in range(len(X)):
    sumX+=X[i]
    meanX=sumX/(len(X))
print("mean of X is",meanX)
for J in range(len(Y)):
    sumY+=Y[J]
    meanY=sumY/(len(Y))
print("mean of Y is",meanY)
numerator=0
denominator=0
for i in range(len(X)):
    numerator+=(X[i]-meanX)*(Y[i]-meanY)
    denominator+=(X[i]-meanX)**2
m=numerator/denominator
print("slope is ",m)
c=meanY-(m*meanX)
print("interceptis ",c)
predicted_values=[(nums*m)+c for nums in X]
print(predicted_values)




```
