```python
("""Interactive NumPy matrix multiplication validator.

This script reads two matrices A and B from the terminal and reports
whether the matrix product A x B is defined (i.e. number of columns
in A equals number of rows in B).

Requirements: Python 3.8+ and NumPy installed. If NumPy is missing the
script prints an instruction to install it.

Usage (PowerShell):
	python "matrixmultiplication.py"

Example input flow:
	Enter rows for matrix A: 2
	Enter columns for matrix A: 3
	Row 1 (3 values): 1 2 3
	Row 2 (3 values): 4 5 6

The script accepts integers or floats for values.
""")

from typing import List
import sys

try:
	import numpy as np
except Exception:
	np = None


def read_int(prompt: str) -> int:
	while True:
		try:
			s = input(prompt).strip()
			return int(s)
		except ValueError:
			print("Please enter a valid integer.")


def read_matrix(name: str) -> List[List[float]]:
	rows = read_int(f"Enter number of rows for matrix {name}: ")
	cols = read_int(f"Enter number of columns for matrix {name}: ")
	if rows <= 0 or cols <= 0:
		print("Rows and columns must be positive.")
		return read_matrix(name)
	mat = []
	for r in range(rows):
		while True:
			s = input(f"Row {r+1} ({cols} values): ").strip()
			parts = s.split()
			if len(parts) != cols:
				print(f"Expected {cols} values, got {len(parts)}. Try again.")
				continue
			row = []
			for p in parts:
				try:
					if "." in p or "e" in p.lower():
						val = float(p)
					else:
						val = int(p)
				except ValueError:
					try:
						val = float(p)
					except ValueError:
						print(f"Could not parse '{p}' as a number. Try the row again.")
						break
				row.append(val)
			else:
				mat.append(row)
				break
	return mat


def can_multiply_numpy(A, B) -> bool:
	if np is None:
		raise RuntimeError("NumPy is not installed. Install with: pip install numpy")
	a = np.array(A)
	b = np.array(B)
	if a.ndim != 2 or b.ndim != 2:
		return False
	return a.shape[1] == b.shape[0]


def main():
	print("NumPy matrix multiplication validator")
	if np is None:
		print("Warning: NumPy is not installed. Install with: pip install numpy")
	A = read_matrix('A')
	B = read_matrix('B')

	# basic shape validation
	cols_A = len(A[0])
	rows_B = len(B)
	defined = False
	try:
		if np is not None:
			defined = can_multiply_numpy(A, B)
		else:
			defined = (cols_A == rows_B)
	except Exception as e:
		# fallback to simple check
		print("NumPy check failed, falling back to basic check:", e)
		defined = (cols_A == rows_B)

	print(f"Matrix A is {len(A)}x{len(A[0])}")
	print(f"Matrix B is {len(B)}x{len(B[0])}")
	if defined:
		print("Result: A x B is defined (multiplication allowed).")
	else:
		print("Result: A x B is NOT defined (columns of A != rows of B).")


if __name__ == '__main__':
	try:
		main()
	except KeyboardInterrupt:
		print("\nInterrupted.")
		sys.exit(1)

```

(ii) Using loops (without using numpy)

```python
"""Pure Python matrix multiplication validator (no NumPy required).

This script reads two matrices A and B from the terminal and reports
whether the matrix product A x B is defined (i.e. number of columns
in A equals number of rows in B).

No external dependencies—uses only Python standard library.

Usage (PowerShell):
    python "matrixmultiplication_pure.py"

Example input flow:
    Enter number of rows for matrix A: 2
    Enter number of columns for matrix A: 3
    Row 1 (3 values): 1 2 3
    Row 2 (3 values): 4 5 6
    Enter number of rows for matrix B: 3
    Enter number of columns for matrix B: 2
    Row 1 (2 values): 7 8
    Row 2 (2 values): 9 10
    Row 3 (2 values): 11 12

The script accepts integers or floats for values.
"""

from typing import List
import sys


def read_int(prompt: str) -> int:
    """Read and validate an integer from user input."""
    while True:
        try:
            s = input(prompt).strip()
            return int(s)
        except ValueError:
            print("Please enter a valid integer.")


def read_matrix(name: str) -> List[List]:
    """Read a matrix from user input.

    Returns a list of rows, where each row is a list of numbers (int or float).
    """
    rows = read_int(f"Enter number of rows for matrix {name}: ")
    cols = read_int(f"Enter number of columns for matrix {name}: ")

    if rows <= 0 or cols <= 0:
        print("Rows and columns must be positive integers.")
        return read_matrix(name)

    matrix = []
    for r in range(rows):
        while True:
            s = input(f"Row {r+1} ({cols} values): ").strip()
            parts = s.split()

            if len(parts) != cols:
                print(f"Expected {cols} values, got {len(parts)}. Try again.")
                continue

            row = []
            valid = True
            for p in parts:
                try:
                    # Try int first
                    if "." not in p and "e" not in p.lower():
                        val = int(p)
                    else:
                        val = float(p)
                except ValueError:
                    print(f"Could not parse '{p}' as a number. Try the row again.")
                    valid = False
                    break
                row.append(val)

            if valid:
                matrix.append(row)
                break

    return matrix


def validate_matrix(matrix: List[List]) -> bool:
    """Check if matrix is valid (non-empty and all rows have same length)."""
    if not matrix or len(matrix) == 0:
        return False
    if not all(isinstance(row, list) for row in matrix):
        return False

    first_row_len = len(matrix[0])
    if first_row_len == 0:
        return False

    return all(len(row) == first_row_len for row in matrix)


def can_multiply(A: List[List], B: List[List]) -> bool:
    """Check if matrix multiplication A x B is defined.

    Returns True if:
    - Both A and B are valid matrices
    - number of columns in A == number of rows in B
    """
    if not validate_matrix(A) or not validate_matrix(B):
        return False

    cols_A = len(A[0])
    rows_B = len(B)
    return cols_A == rows_B


def display_matrix(matrix: List[List], name: str):
    """Pretty-print a matrix."""
    print(f"\nMatrix {name}:")
    for row in matrix:
        print("  [", " ".join(f"{val:8.2f}" if isinstance(val, float) else f"{val:8}" for val in row), "]")


def main():
    print("=" * 60)
    print("Pure Python Matrix Multiplication Validator")
    print("=" * 60)
    print()

    A = read_matrix('A')
    print()
    B = read_matrix('B')
    print()

    # Validate and check
    if not validate_matrix(A):
        print("Error: Matrix A is invalid.")
        return
    if not validate_matrix(B):
        print("Error: Matrix B is invalid.")
        return

    display_matrix(A, "A")
    display_matrix(B, "B")
    print()

    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])

    print(f"Matrix A dimensions: {rows_A} x {cols_A}")
    print(f"Matrix B dimensions: {rows_B} x {cols_B}")
    print()

    is_valid = can_multiply(A, B)

    if is_valid:
        print("✓ Result: A x B IS DEFINED")
        print(f"  Multiplication is valid because:")
        print(f"  - Number of columns in A ({cols_A}) == Number of rows in B ({rows_B})")
        print(f"  - The product will be a {rows_A} x {cols_B} matrix")
    else:
        print("✗ Result: A x B IS NOT DEFINED")
        print(f"  Multiplication is invalid because:")
        print(f"  - Number of columns in A ({cols_A}) != Number of rows in B ({rows_B})")

    print()


if __name__ == '__main__':
    try:
        main()
    except KeyboardInterrupt:
        print("\nInterrupted by user.")
        sys.exit(1)
    except Exception as e:
        print(f"Error: {e}")
        sys.exit(1)
```


Compute:

No libraries allowed.(sklearn or numpy)


```python
"""Pure Python Linear Regression Calculator (no libraries).

Computes mean, slope, intercept, and predicted values for simple linear regression.
Formula: y = mx + c
where m = slope, c = intercept
"""


def mean(data):
    """Calculate the mean of a list of numbers."""
    return sum(data) / len(data)


def calculate_slope_and_intercept(X, y):
    """Calculate slope (m) and intercept (c) for linear regression.

    Using the formulas:
    m = Σ((x - mean_x) * (y - mean_y)) / Σ((x - mean_x)²)
    c = mean_y - m * mean_x
    """
    n = len(X)
    mean_x = mean(X)
    mean_y = mean(y)

    # Calculate numerator and denominator for slope
    numerator = sum((X[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    denominator = sum((X[i] - mean_x) ** 2 for i in range(n))

    # Avoid division by zero
    if denominator == 0:
        raise ValueError("Cannot calculate slope: denominator is zero")

    m = numerator / denominator
    c = mean_y - m * mean_x

    return m, c


def predict(X, m, c):
    """Predict y values using the linear equation y = mx + c."""
    return [m * x + c for x in X]


def print_results(X, y, mean_x, mean_y, m, c, y_pred):
    """Pretty-print all results."""
    print("=" * 70)
    print("LINEAR REGRESSION ANALYSIS (Pure Python, No Libraries)")
    print("=" * 70)
    print()

    print("INPUT DATA:")
    print(f"  X values: {X}")
    print(f"  y values: {y}")
    print()

    print("CALCULATED VALUES:")
    print(f"  Mean of X:  {mean_x:.4f}")
    print(f"  Mean of y:  {mean_y:.4f}")
    print()

    print("LINEAR REGRESSION (y = mx + c):")
    print(f"  Slope (m):      {m:.4f}")
    print(f"  Intercept (c):  {c:.4f}")
    print()

    print("PREDICTED VALUES:")
    print(f"  {'X':>5} | {'Actual y':>10} | {'Predicted y':>12} | {'Error':>8}")
    print("  " + "-" * 50)
    for i in range(len(X)):
        error = y[i] - y_pred[i]
        print(f"  {X[i]:>5} | {y[i]:>10} | {y_pred[i]:>12.4f} | {error:>8.4f}")

    print()
    print("EQUATION: y = {:.4f}x + {:.4f}".format(m, c))
    print("=" * 70)


def main():
    # Given data
    X = [1, 2, 3, 4, 5]
    y = [2, 4, 5, 4, 5]

    # Compute mean of X and y
    mean_x = mean(X)
    mean_y = mean(y)

    # Compute slope and intercept
    m, c = calculate_slope_and_intercept(X, y)

    # Compute predicted values
    y_pred = predict(X, m, c)

    # Display results
    print_results(X, y, mean_x, mean_y, m, c, y_pred)

    # Return results for verification
    return {
        'mean_X': mean_x,
        'mean_y': mean_y,
        'slope': m,
        'intercept': c,
        'predicted': y_pred
    }


if __name__ == '__main__':
    results = main()

    # Summary
    print("\nQUICK REFERENCE:")
    print(f"  Mean of X:     {results['mean_X']:.4f}")
    print(f"  Mean of y:     {results['mean_y']:.4f}")
    print(f"  Slope (m):     {results['slope']:.4f}")
    print(f"  Intercept (c): {results['intercept']:.4f}")
    print(f"  Predicted y:   {[round(p, 4) for p in results['predicted']]}")
```
