```python
X = [1,2,3,4,5]
Y = [2,4,5,4,5]

# MEAN OF X

meanX = sum(X)/len(X)
print(meanX)
```

```python
# MEAN OF Y

X = [1,2,3,4,5]
Y = [2,4,5,4,5]

meanY = sum(Y)/len(Y)
print(meanY)
```

```python
# SLOPE(m)

X = [1,2,3,4,5]
Y = [2,4,5,4,5]
num = 0
den = 0

for i in range(len(X)):
  num += (X[i] - meanX) * (Y[i] - meanY)
  den += (X[i] - meanX) ** 2
if (den == 0):
    print("SLOPE IS NOT DEFINED i.e.. THE LINE IS PARALLEL TO Y AXIS")
else:
    m = num / den
    print(m)
```

```python
# INTERCEPT(b)

b = meanY - (m * meanX)
print(b)
```
