## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,100)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
y = x*2
plt.plot(x,y,'g',linewidth = 1, label = 'x*2')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('x')
ax1.set_ylabel('z')
plt.plot(x,z,'b',label='z=x**2')

ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
plt.plot(x,y,'b',label='y=x*2')
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))
y = 2*x
ax1.plot(x,y,'b-',linestyle=':',linewidth= 2, label = 'y = 2x')
z = x**2
ax2.plot(x,z,'r-',linewidth= 1, label = 'z = x^2')
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,2))
y = 2*x
ax1.plot(x,y,'b-',linewidth= 2, label = 'y = 2x')
z = x**2
ax2.plot(x,z,'r-',linestyle=':',linewidth= 2, label = 'z = x^2')
```
