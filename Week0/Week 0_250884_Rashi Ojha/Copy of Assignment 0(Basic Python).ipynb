**Q.4**
(i) Given:

```python
X = [1,2,3,4,5]
y = [2,4,5,4,5]
```

Compute:

mean of X

mean of y

slope (m)

intercept (c)

predicted values

No libraries allowed.(sklearn or numpy)


```python
x = [1,2,3,4,5]
y = [2,4,5,4,5]
mean_x= sum(x)/len(x)
mean_y=sum(y)/len(y)
print("The mean of x is",mean_x)
print("The mean of y is",mean_y)
numerator = 0
denominator = 0

for i in range(len(x)):
    numerator += (x[i] - mean_x) * (y[i] - mean_y)
    denominator += (x[i] - mean_x) ** 2

m = numerator / denominator
print("The slope is",m)

c=(mean_y)-(m*mean_x)
print("The intercept is",c)
```
