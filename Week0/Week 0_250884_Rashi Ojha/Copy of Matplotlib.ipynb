**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt

import numpy as np

x = np.arange(0,100)
y = x*2

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color="blue")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("title")
plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt

plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt

fig, axes= plt.subplots(nrows=1 , ncols = 2)
axes[0].plot(x, y, linestyle='--', color='blue')
axes[0].set_title('(x,y)')

axes[1].plot(x, z,color='red')
axes[1].set_title('(x,z)')

fig.tight_layout()
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt

fig, axes= plt.subplots(nrows=1 , ncols = 2, figsize=(10,2))
axes[0].plot(x, y, linestyle='--', color='blue')
axes[0].set_title('(x,y)')

axes[1].plot(x, z,color='red')
axes[1].set_title('(x,z)')

fig.tight_layout()
plt.show()
```

# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
values = [8, 15, 22, 30]

#1.Histogram
plt.figure()
plt.hist(data, bins=30)
plt.title('Histogram Example')
plt.show()

#2.Scatter Plot
plt.figure()
plt.scatter(data[:10], data[10:20])
plt.title('Scatter Plot Example')
plt.show()

#3.Boxplot
plt.figure()
plt.boxplot([data[:50], data[50:100]])
plt.title('Boxplot Example')
plt.show()

#4.Pie charts
plt.figure()
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```
